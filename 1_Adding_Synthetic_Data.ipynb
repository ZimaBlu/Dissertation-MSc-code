```python
import pandas as pd
import numpy as np
```

```python
hr = pd.read_csv("hr.csv")
sm = pd.read_csv("sedentary_minutes.csv")
sss = pd.read_csv("sleep_score_sleep.csv")
s = pd.read_csv("steps.csv")
vam = pd.read_csv("very_active_minutes.csv")
```

```python
sm.drop(sm.columns[0], inplace = True, axis=1)
sss.drop(sss.columns[0], inplace = True, axis=1)
s.drop(s.columns[0], inplace=True, axis=1)
vam.drop(vam.columns[0], inplace = True, axis=1)
```

```python
df = hr.set_index('date').join(sm.set_index('date')).join(sss.set_index('date')).join(s.set_index('date')).join(vam.set_index('date'))
```

```python
df['Fecal calprotectin >100'] = np.random.choice([1, 0], len(df), p=[0.15, 0.85])
df['C-reactive protein >10'] = np.random.choice([1, 0], len(df), p=[0.15, 0.85])
df['Hospital admission'] = np.random.choice([1, 0], len(df), p=[0.1, 0.9])
df['Prescribed prednisolone'] = np.random.choice([1, 0], len(df), p=[0.3, 0.7])
df['Change of medication'] = np.random.choice([1, 0], len(df), p=[0.1, 0.9])
```

```python
df['Total indicators'] = df['Fecal calprotectin >100'] + df['C-reactive protein >10'] + df['Hospital admission'] + df['Prescribed prednisolone'] + df['Change of medication']
```
